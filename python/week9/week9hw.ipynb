```python
# Dice Rolling Simulator. Use the random module to create a function that simulates rolling two six-sided dice and returns their sum.
# Then, create a program that asks the user how many times to roll the dice and uses a dictionary to track and display the frequency of each sum.

from mymodule import rolling_simmulator as rs

roll_cnt = 5
max_cnt = 100
res = {}

def check_user_input():
    global roll_cnt
    while True:
        a = input(f"Enter how many times to roll dice or just push 'Enter' to roll {roll_cnt} times by default.")
        if a == "":
            print(f"We will roll dice {roll_cnt} times by default")
            return roll_cnt
            break
        elif a.isdigit():
            a = int(a)
            if a <= max_cnt:
                roll_cnt = a
                print(f"We will roll dice {roll_cnt} times as you want")
                break
            else:
                print(f"You should enter digit from 1 to {max_cnt}")
        else:
            print("You should enter digit")
    return (roll_cnt)

def main():
    print("Welcome to Dice Rolling game!")
    roll_cnt = check_user_input()
    for i in range(roll_cnt):
        roll = rs()
        # print(roll)
        if roll[2] in res:
            res[roll[2]] += 1
        else:
            res[roll[2]] = 1
    # print(res)
    print(dict(sorted(res.items(), key = lambda x: x[1], reverse = True)))

if __name__ == "__main__":
    main()
```
