```python
#!/usr/bin/python3

# Modify your To-Do List program from earlier. When the program starts, it should read any existing tasks from a file named tasks.txt.
# When the user adds or deletes a task, the program should update the file. This way, the tasks are not lost when the program closes.

# tasks = ["Math", "Python", "CompSys", "OfficeHours", "Games", "HomeWork", "Test"]
errmsg = "\n You entered wrong value, please try again\n"
emptymsg = "Tasks list is empty"
file_path = r"C:\Users\NRG\GH\python\week8hw\todo_list.txt"

def open_file(edit_type, tasks = None):
    try:
        with open(file_path, edit_type) as f:
            if edit_type == "r":
                return f.read().splitlines()
            elif edit_type == "w":
                if tasks:
                    f.write("\n".join(tasks))
    except FileNotFoundError:
        open(file_path, "w").close()


def checkTaskExist(tasks, taskName):
    for i, task in enumerate(tasks):
        if task.lower() == taskName.lower():
            return i
    return -1

def addTasks(tasks):
    print("\nYou have selected the operation - add tasks:")
    i = 0
    if tasks:
        i = len(tasks)
    j = i
    while True:
        a = input("Add your task # %d or press enter for quit:" % (i + 1))
        if a == "":
            break
        else:
            b = checkTaskExist(tasks, a)
            if  b == -1:
                tasks.append(a)
                i += 1
            else:
                print("You enterd task - %s that exist in your task list by number %d" % (a, b + 1))
    if i - j > 0:
        print("Great! You added %d tasks:" % (i - j))
        for i, task in enumerate(tasks[j:], start = j + 1):
            print(" Task # %d is: %s" % (i, task))
    return tasks

def printTasks(tasks):
    print("\nYou have selected the operation - view all tasks:")
    if tasks:
        for i, task in enumerate(tasks, start = 1):
            print("Task # %d is - %s" %(i, task))
        return
    else:
        print(emptymsg)

def deleteTasks(tasks):
    if tasks:
        print("\nYou have selected the operation - delete tasks:")
        while True:
            choice = input("Select how to delete task:\n1. By number\n2. By name\nEnter '1' or '2'\n or press enter for to go to main menu: ")
            if choice == "":
                return tasks
            else:
                try:
                    choice = int(choice)
                    if choice in [1, 2]:
                        break
                    else:
                        print(" You should enter only '1' or '2', try again")
                except ValueError:
                    print(errmsg)

        tasksForDel = []
        i = 0
        j = 0
        if choice == 1:
            while True:
                a = input("Enter # of task (less than %d) for deleting and press enter \nOr press enter for quit:" % (len(tasks)+1))
                if a == "":
                    break
                elif a.isdigit():
                    if int(a) <= len(tasks):
                        tasksForDel.append(int(a) - 1)
                        i += 1
                    else:
                        print("You entered # that doesn't exist in your tasks list, it should be less than %d, enter again" % (len(tasks) + 1))
                elif a.isdigit() == False:
                    print(errmsg)
        else:
            while True:
                a = input("Enter name of task for deleting and press enter \nOr press enter for quit:")
                if a == "":
                    break
                else:
                    b = checkTaskExist(tasks, a)
                    if b != -1:
                        tasksForDel.append(b)
                        i += 1
                    else:
                        print("You entered task that doesn't exist in your tasks list, enter again")

        tasksForDel = list(set(tasksForDel))  

        tasksForDel.sort(reverse=True)
        removedTasks = []
        for i in tasksForDel:
            a = tasks.pop(i)
            removedTasks.append((i, a))

        if len(removedTasks) > 0:
            print("Great! You removed next tasks:")
            removedTasks.sort()
            for i, task in removedTasks:
                print(" Task # %d is: %s" % (i + 1, task))
            if len(tasks) == 0:
                print(emptymsg)
        return tasks
    else:
        print(emptymsg)


def main():
    while True:
        oper = input("Enter number of operation you would like:\n 1 - Add task\n 2 - View all tasks\n 3 - Delete task by its name or number\n or press enter for quit")
        if oper == "":
            print("You've finished your session, Bye!")
            break
        elif oper == "1":
            tasks = addTasks(open_file("r"))
            open_file("w", tasks)
        elif oper == "2":
            tasks = open_file("r")
            printTasks(tasks)
        elif oper == "3":
            tasks = deleteTasks(open_file("r"))
            open_file("w", tasks)
        else:
            print(errmsg)

if __name__ == '__main__':
    main()
```
