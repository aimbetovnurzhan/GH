```python
# Project:
# Simple Bank Account Class. Create a BankAccount class. Each account should have an owner name and a balance.
# The class should have methods to deposit() money, withdraw() money (don't allow withdrawal if funds are insufficient), and display_balance().

class BankAccount():
    owner = ""
    balance = 0

    def __init__(self, name, amount):
        self.owner = name
        self.balance = amount
    
    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            print(f"{self.owner} - you don't have enough money to wihdraw ${amount}, your current balance is - ${self.balance}")
    
    def printInfo(self):
        print(f"Owner - {self.owner}, current balance is - ${self.balance}")

account1 = BankAccount("Nurzhan", 1000)
account2 = BankAccount("Aliya", 10000)

account1.deposit(500)
account2.withdraw(4000)

account1.printInfo()
account2.printInfo()

account1.withdraw(2000)
```
