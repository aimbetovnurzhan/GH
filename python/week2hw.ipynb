```python
#Сделать калькулятор, ввести общую сумму чека, кол-во людей (имена людей) и сумму чаевых.
#В итоге нужно вернуть  тотал сумму чека, сумму чаевых.
# Дальше разделить чек по именам с суммами на каждого, чаевые и тотал на каждого.

bilAmnt = float(input("Введите сумму чека:"))
errmsg = "\nОшбика ввода. Введите заново\n"
while True:
    cnt = int(input("Количество гостей:"))
    if cnt > 0:
        break
    else:
        print(errmsg)

eachBillAmnt = bilAmnt/cnt
totalTips = 0
tipsList=[]

if cnt > 1:
    while True:
        choice1 = int(input("Как будут распределены чаевые:\n1. Общий на всех.\n2. Каждый отдельно.\nВведите '1' или '2': "))
        if choice1 == 1 or choice1 == 2:
            break
        else:
            print(errmsg)

if choice1 == 1 or cnt == 1:
    while True:
        choice2 = int(input("Выберите метод ввода чаевых:\n1. Ручной ввод суммы чаевых.\n2. Процент от суммы чека.\nВведите '1' или '2': "))
        if choice2 == 1 or choice2 == 2:
            break
        else:
            print(errmsg)
    if choice2 == 1:
        totalTips = float(input("Введите сумму чаевых:"))
    if choice2 == 2:
        tipsFraction = int(input("Введите процент для чаевых:"))
        totalTips = tipsFraction * bilAmnt / 100
    for i in range(cnt):
        tipsList.append(totalTips/cnt)

if choice1 == 2:
    for i in range(cnt):
        while True:
            choice2 = int(input("Гость №%d Выберите метод ввода чаевых:\n1. Ручной ввод суммы чаевых.\n2. Процент от суммы чека.\nВведите '1' или '2': " % (i + 1)))
            if choice2 == 1 or choice2 == 2:
                break
            else:
                print(errmsg)
        if choice2 == 1:
            tips = float(input("Гость №%d введите сумму чаевых:" % (i + 1)))
        if choice2 == 2:
            tipsFraction = int(input("Гость №%d введите процент для чаевых:" % (i + 1)))
            tips = eachBillAmnt * tipsFraction / 100
        tipsList.append(tips)
        totalTips = sum(tipsList)

if cnt > 1:
    for i in range(cnt):
        print('Гость № %d.\n Доля суммы к оплате: $%0.2f.\n Tips: $%0.2f.\n Общая сумма к оплате: $ %0.2f.\n' %  (i+1, eachBillAmnt, tipsList[i], eachBillAmnt + tipsList[i]))
print('TotalTips: $%0.2f.\nОбщая сумма к оплате: $ %0.2f.' %  (totalTips, bilAmnt + totalTips))
```

```python
# A. match_ends
# Given a list of strings, return the count of the number of
# strings where the string length is 2 or more and the first
# and last chars of the string are the same.
# Note: python does not have a ++ operator, but += works.
def match_ends(words):
    # +++your code here+++
    j = 0
    for i in range(len(words)):
        if len(words[i]) >= 2 and words[i][0] == words[i][-1]:
            j += 1
    return j

# B. front_x
# Given a list of strings, return a list with the strings
# in sorted order, except group all the strings that begin with 'x' first.
# e.g. ['mix', 'xyz', 'apple', 'xanadu', 'aardvark'] yields
# ['xanadu', 'xyz', 'aardvark', 'apple', 'mix']
# Hint: this can be done by making 2 lists and sorting each of them
# before combining them.
def front_x(words):
    # +++your code here+++
    k, l = 0, 0
    n = len(words)
    for k in range(n-1):
        for l in range(k+1, n):
            if words[k] > words[l]:
                words[k], words[l] = words[l], words[k]
    l = 0
    for k in range(n):
        if words[k].startswith("x"):
            words.insert(0+l,words[k])
            l += 1
            words.pop(k+1)
    return words

# C. sort_last
# Given a list of non-empty tuples, return a list sorted in increasing
# order by the last element in each tuple.
# e.g. [(1, 7), (1, 3), (3, 4, 5), (2, 2)] yields
# [(2, 2), (1, 3), (3, 4, 5), (1, 7)]
# Hint: use a custom key= function to extract the last element form each tuple.
def sort_last(tuples):
    # +++your code here+++
    n = len(tuples)
    for i in range(n):
        for j in range(i+1,n):
            if tuples[i][-1] > tuples[j][-1]:
                tuples[i], tuples[j] = tuples[j], tuples[i]
    return tuples


# Simple provided test() function used in main() to print
# what each function returns vs. what it's supposed to return.
def test(got, expected):
    if got == expected:
        prefix = ' OK '
    else:
        prefix = '  X '
    print('%s got: %s expected: %s' % (prefix, repr(got), repr(expected)))


# Calls the above functions with interesting inputs.
def main():
  print('match_ends')
  test(match_ends(['aba', 'xyz', 'aa', 'x', 'bbb']), 3)
  test(match_ends(['', 'x', 'xy', 'xyx', 'xx']), 2)
  test(match_ends(['aaa', 'be', 'abc', 'hello']), 1)

  print()
  print('front_x')
  test(front_x(['bbb', 'ccc', 'axx', 'xzz', 'xaa']),
       ['xaa', 'xzz', 'axx', 'bbb', 'ccc'])
  test(front_x(['ccc', 'bbb', 'aaa', 'xcc', 'xaa']),
       ['xaa', 'xcc', 'aaa', 'bbb', 'ccc'])
  test(front_x(['mix', 'xyz', 'apple', 'xanadu', 'aardvark']),
       ['xanadu', 'xyz', 'aardvark', 'apple', 'mix'])


  print()
  print('sort_last')
  test(sort_last([(1, 3), (3, 2), (2, 1)]),
       [(2, 1), (3, 2), (1, 3)])
  test(sort_last([(2, 3), (1, 2), (3, 1)]),
       [(3, 1), (1, 2), (2, 3)])
  test(sort_last([(1, 7), (1, 3), (3, 4, 5), (2, 2)]),
       [(2, 2), (1, 3), (3, 4, 5), (1, 7)])

if __name__ == '__main__':
    main()
```
