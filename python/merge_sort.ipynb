```python
def merge_sort(arr):
    # если в списке 0 или 1 элемент — он уже отсортирован
    if len(arr) < 2:
        return arr

    mid = len(arr) // 2  # середина списка
    left = merge_sort(arr[:mid])   # сортируем левую половину
    right = merge_sort(arr[mid:])  # сортируем правую половину

    # сливаем две отсортированные половины
    return merge(left, right)


def merge(left, right):
    result = []  # сюда складываем отсортированные элементы
    i = j = 0    # индексы для left и right

    # пока в обеих половинах есть элементы
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # добавляем "хвост" — что осталось в одной из половинок
    result.extend(left[i:])
    result.extend(right[j:])
    return result


nums = [5, 2, 9, 1, 7, 12, 19, 11, 17]#, 22, 29, 21, 27, 212, 219, 211, 217, 32, 39, 31, 37, 312, 319, 311, 317, 322, 329, 321, 327, 3212, 3219, 3211, 3217]
sorted_nums = merge_sort(nums)
print(sorted_nums)  # [1, 2, 5, 7, 9]
```
