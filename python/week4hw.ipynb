```python
#Сделать калькулятор, ввести общую сумму чека, кол-во людей (имена людей) и сумму чаевых.
#В итоге нужно вернуть  тотал сумму чека, сумму чаевых.
# Дальше разделить чек по именам с суммами на каждого, чаевые и тотал на каждого.

errmsg = "\nValue error, try again\n"

def checkInteger(max=None):
    if max is None:
        max = 100
    while True:
        a = input()
        try:
            val = int(a)
            if val > 0:
                if  val <= max:
                    return val
                else:
                    print("Should enter a number less than or equal to %d." % max)
            else:
                print("Should enter a positive number.")
        except ValueError:
            print(errmsg)

def checkFloat(max=None):
    if max is None:
        max = 10000.0
    while True:
        a = input()
        try:
            val = float(a)
            if val > 0.0:
                if  val <= max:
                    return val
                else:
                    print("Should enter a number less than or equal to %0.2f." % max)
            else:
                print("Should enter a positive number.")
        except ValueError:
            print(errmsg)

print("Please enter the bill amount:")
bilAmnt = checkFloat()

print("Please enter the number of guests:")
cnt = checkInteger()

eachBillAmnt = bilAmnt/cnt
totalTips = 0
tipsList=[]

if cnt > 1:
    print("How do you want to pay tips? \n1. Together.\n2. Separately.\nEnter '1' or '2': ")
    choice1 = checkInteger(2)

if choice1 == 1 or cnt == 1:
    print("How do you want to pay tips? \n1. Enter amount manually.\n2. Enter percentage.\nEnter '1' or '2': ")
    choice2 = checkInteger(2)
    if choice2 == 1:
        print("Enter total tips amount:")
        totalTips = checkFloat(bilAmnt/2)
    if choice2 == 2:
        print("Enter percentage for tips:")
        tipsFraction = checkFloat(50)
        totalTips = tipsFraction * bilAmnt / 100
    for i in range(cnt):
        tipsList.append(totalTips/cnt)

if choice1 == 2:
    for i in range(cnt):
        print("Guest №%d how do you want to pay tips? \n1. Enter amount manually.\n2. Enter percentage.\nEnter '1' or '2': " % (i + 1))
        choice2 = checkInteger(2)
        if choice2 == 1:
            print("Guest №%d enter tips amount:" % (i + 1))
            tips = checkFloat(eachBillAmnt/2)
        if choice2 == 2:
            print("Guest №%d enter percentage for tips:" % (i + 1))
            tipsFraction = checkFloat(50)
            tips = eachBillAmnt * tipsFraction / 100
        tipsList.append(tips)
        totalTips = sum(tipsList)

if cnt > 1:
    for i in range(cnt):
        if tipsList[i]/eachBillAmnt > 0.1:
            farewellPhrase = "You are very generous! We will be glad to see you again, Thank you!"
        else:
            farewellPhrase = "Thank you for your visit! We hope you won't come back!"
        print("\nGuest № %d.\n Bill share: $%0.2f.\n Tips: $%0.2f.\n Total to pay: $ %0.2f.\n %s\n" %  (i + 1, eachBillAmnt, tipsList[i], eachBillAmnt + tipsList[i], farewellPhrase))

print('Total Tips: $%0.2f.\nTotal amount for pay: $ %0.2f.' %  (totalTips, bilAmnt + totalTips))
```
