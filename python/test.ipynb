```python
n = 5

twos = n//2
ones = n%2
res = []
sublist = []
while twos > 0 or ones <= n:
    for i in range(twos):
        sublist.append(2)
    for i in range(ones):
        sublist.append(1)
    # print(sublist)
    if twos*2 == n or ones == n: 
        res.append(sublist)
    else:
        res += get_permutations(sublist)
    sublist = []
    twos -= 1
    ones += 2

# print(res)
unique_variations = list(set([tuple(x) for x in res]))
print(len([list(x) for x in unique_variations]))


def get_permutations(nums):
    result = []
    
    def backtrack(start):
        if start == len(nums):
            result.append(nums[:])
            return
        
        for i in range(start, len(nums)):
            nums[start], nums[i] = nums[i], nums[start]
            backtrack(start + 1)
            nums[start], nums[i] = nums[i], nums[start]
            
    backtrack(0)
    return result
```
