```python
# Simple Word Counter. Write a program that asks the user for a sentence. 
# The program should then count the occurrences of each word in the sentence
# and display the result as a dictionary.

# sentence = input("Enter a sentence: ")
sentence = "This is a test sentence. Sentence should be use only for test. This test is simple, a simple test.2 3 4 7 1 9 8 6 5 4 3 6 3 2 3 4 6 5"
cleanedSentence = sentence.replace('.', '').replace(',', '')  # Remove punctuation
words = cleanedSentence.split()
loweredWords = [word.lower() for word in words]  # Convert all words to lowercase
word_count = {}

for word in loweredWords:
    count = loweredWords.count(word)
    if word not in word_count:
        word_count[word] = count
print(word_count)
```
